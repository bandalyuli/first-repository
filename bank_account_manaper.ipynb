```python
# 1. account 클래스 만들기
from random import *
class Account:
  account_num = 0
  def __init__(self, account_holder):
    self.bank_name = '모두은행'  # 은행이름
    self.account_holder = account_holder # 예금주
    self.account_number = self.create_account_number() # 계좌번호 생성
    self.balance = 0 # 잔액
    self.deposit_count = 0  # 입금 횟수 - 입금 5회 이자 생성을 위해 입금만 세기
    Account.account_num += 1 # 계좌 수 늘렸을 때
    self.account_history = '' # 입출금 기록
    self.total_count = 0 # 입출금(거래) 횟수 - 입금 5회 이자 생성을 위해 따로 만들고 출금 횟수도 함께

  def create_account_number(self): #계좌번호 생성
    part1 = randint(100, 999) # 3자리
    part2 = randint(10,99) # 2자리
    part3 = randint(100000, 999999) # 6자리
    return f'{part1}-{part2}-{part3}'
    # 한 줄로 한다면? return f'{randint(100, 999)}-{randint(10,99)}-{randint(100000, 999999)}'

  # 2. 입금 메서드
  def deposit(self, cost): #cost 는 입금액
    if cost < 1: # 입금액이 1원 미만일 때
       print("입금 금액은 1원 이상이어야 합니다.")
       return
    else: # 입금액이 1원 이상일 때
      self.balance += cost # 잔액 변동사항 값
      self.deposit_count += 1 # 입금 횟수 추가
      self.total_count += 1 # 거래 횟수 추가
      self.account_history += f'{self.total_count}회: 입금\t금액: {cost}\t잔액: {self.balance}\n'
        #

      if self.deposit_count % 5 == 0: # 입급회수 5회마다
        self.deposit_count += 1
        self.total_count += 1
        interest = self.balance * 0.01 # 1퍼센트 이자
        self.balance += interest
        self.account_history += f'{self.total_count}회: 이자\t금액: {interest}\t잔액: {self.balance}\n'


  # 3. 출금 메서드
  def withdraw(self, cost):
    if cost > self.balance:
      print('계좌의 잔액이 모자랍니다. 잔고 이하의 금액을 입력해주세요.')
    else:
      self.balance -= cost
      self.total_count += 1
      self.account_history += f'{self.total_count}회: 출금\t금액: {cost}\t잔액: {self.balance}\n'
  # 입출금 출력을 count 속성을 2개로 나눠서 진행
  # 다른 방법은 히스토리를 account와 withdraw를 나눠서 히스토리에 합치는 걸로


  #4. get_account_num() 메서드 - # 계좌 수
  def get_account_num(self):
    print(Account.account_num)

  #5. history() 메서드 #입출금 기록
  def history(self):
    print(self.account_history)
```

```python
my_account = Account('유진국')
```
